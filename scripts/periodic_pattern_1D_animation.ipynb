## Spectrum generation with known peaks

```python
fig = plt.figure(num=None, figsize=(16, 8), dpi=fig_dpi, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.family':'Source Sans Pro'})

t = np.linspace(-L/2, L/2, L)

ax1 = plt.subplot(111)

line, = ax1.plot([], [], color='blue', linewidth=2, zorder=3)
line2, = ax1.plot([], [], color='blue', linewidth=2, zorder=3)
line3, = ax1.plot([], [], color='blue', linewidth=2, zorder=3)

def init():
    line2.set_data([], [])
    line3.set_data([], [])
    return line2, line3

def animate(i):
    fc = 0.01
    x_shift = i/5
    pattern_1D = 0.5 * (1 + np.cos(2 * pi * fc *(t - x_shift)))
    
    x_0 = x_shift%(1/fc)

    Y = np.fft.fft(pattern_1D)

    P2 = np.abs(Y/L)
    P1 = P2.copy()

    P1[:int(L/2)] = P2[int(L/2):]
    P1[int(L/2):] = P2[:int(L/2)]

    P1 *= 2

    f = np.arange(-int(L/2), int(L/2))/(L)

    line.set_data([0, 0], [0, 1])
    line2.set_data([L*fc, L*fc], [0, 0.5])
    line3.set_data([-L*fc, -L*fc], [0, 0.5])
    
    return line, line2, line3

ax1.set_xlim(-L/4, L/4)
ax1.set_ylim(0, 1.2)
ax1.set_title('(c) Intensité du spectre', fontsize=25)
ax1.set_xticks([0])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ani = animation.FuncAnimation(fig, animate, init_func=init, frames=500, blit=True, interval=20, repeat=False)
ani.save('slide1/signal_1D_sinewave_fixed_f_moving_phi_spectrum.mp4')
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(16, 3), dpi=fig_dpi, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.family':'Source Sans Pro'})
props = dict(boxstyle='square', facecolor='white', alpha=0.5)

ax1 = plt.subplot(111)


L_tot = 2048
fc = 0.02
t = np.arange(-int(L_tot/2), int(L_tot/2))
f = np.arange(-int(L_tot/2), int(L_tot/2))/(L_tot)

line, = ax1.plot([], [], color='blue', linewidth=2, zorder=3)
line2, = ax1.plot([], [], '--r', linewidth=2, zorder=3)

text_slope = ax1.text(600, 50, '', color='red', fontsize=14, verticalalignment='top', bbox=props)
textstr = ''

def init():
    line.set_data([],[])
    return line,

def animate(i):
    phi_0 = i/20
    pattern_1D = 0.5 * (1 + np.cos(2 * pi * fc * t + phi_0))
    Y = np.fft.fft(pattern_1D)
    Y = np.fft.fftshift(Y)
    P2 = np.abs(Y/L_tot)*2
    
    pic_x = (pattern_1D.shape[0]*fc)

    gaussian_filter = np.exp(-((t-pic_x))**2)

    filtered_spectrum = gaussian_filter * Y
    filtered_spectrum = np.fft.fftshift(filtered_spectrum)
    ift = np.fft.ifft(filtered_spectrum)
    wrapped_phase = np.angle(ift)
    unwrapped_phase = np.unwrap(wrapped_phase) - (L_tot * fc * pi)

    line.set_data(t, unwrapped_phase)

    x_vec = t.reshape((-1,1))
    y_vec = unwrapped_phase

    model = LinearRegression().fit(x_vec, y_vec)
    r_sq = model.score(x_vec, y_vec)
    # print('coefficient of determination:', r_sq)
    # print('intercept:', model.intercept_)
    # print('slope:', model.coef_)

    y_coeff = model.coef_
    y_offset = model.intercept_

    y_reg = t*y_coeff + y_offset

    line2.set_data(t, y_reg)

    if y_offset < 0:
        textstr = r'$\phi=%.2f x %.2f$' % (y_coeff,y_offset)
    else:
        textstr = r'$\phi=%.2f x + %.2f$' % (y_coeff,y_offset)

    text_slope.set_text(textstr)

    return line


ax1.grid(zorder=3)
ax1.set_title('(c) Phase enroulée du pic fréquentiel sélectionné', fontsize=25)
ax1.set_xlim(-L_tot/2, L_tot/2)
ax1.set_ylim(-150, 150)
ax1.set_xticks([-1024, 0, 1024])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ani = animation.FuncAnimation(fig, animate, init_func=init, frames=400, blit=True, interval=20, repeat=False)
# ani.save('../../oral_presentation/GIFS/methode_1D/sinewave_1D_wrapped_phase.gif', writer='pillow', fps=20)
ani.save('sinewave_1D_unwrapped_phase.mp4')
plt.show()
```
